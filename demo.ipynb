```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scripts.invariants import InvariantsCalculator
```

```python
pic = Image.open("data/lena.png")
plt.imshow(pic, cmap='gray')
plt.show()
invar = InvariantsCalculator()
invar.calculateInvariants(np.asmatrix(np.array(pic)))
print('Representation: {}'.format(invar.getInvariants()))
```

```python
img_rotated_90 = pic.rotate(90)
plt.imshow(img_rotated_90, cmap='gray')
plt.show()
invar90 = InvariantsCalculator()
invar90.calculateInvariants(np.asmatrix(np.array(img_rotated_90)))
print('Representation: {}'.format(invar90.getInvariants()))
```
